```python
data = pd.read_csv('adult.csv')

#columns with null values are workclass,occupation and native-country

#fill the three columns with their respective mode values
data['workclass'].fillna(data['workclass'].mode()[0],inplace=True)

data['occupation'].fillna(data['occupation'].mode()[0],inplace=True)

data['native-country'].fillna(data['native-country'].mode()[0],inplace=True)

#check if there is still null values
data.isna().sum()
```

```python
############# WRITE YOUR CODE IN THIS CELL (IF APPLICABLE)####################

#check the values in income columns
data['income'].unique()

#encode the values as 0,1,2,3 respectively

data['income'].replace('<=50K',0,inplace=True)

data['income'].replace('>50K',1,inplace=True)

data['income'].replace('<=50K.',2,inplace=True)

data['income'].replace('>50K.',3,inplace=True)

#display the result
data['income'].unique()
```

```python
#check the unique value in the education column

data['education'].unique()
```

```python
#create the mapped value

map_values = {
    'Preschool': 0,
    '1st-4th': 1,
    '5th-6th': 2,
    '7th-8th': 3,
    '9th': 4,
    '10th': 5,
    '11th': 6,
    '12th': 7,
    'HS-grad': 8,
    'Some-college': 9,
    'Assoc-acdm': 10,
    'Assoc-voc': 11,
    'Bachelors': 12,
    'Masters': 13,
    'Doctorate': 14,
    'Prof-school': 15
}

# encode the education column with the mapped values
data['education'] = data['education'].map(map_values)
```

```python
#display the unique values in the sex column

data['sex'].unique()
```

```python
#encode the sex column as the nominal column

data['sex'].replace('Male',0,inplace=True)

data['sex'].replace('Female',1,inplace=True)
```

```python
#import a normaizer function
from sklearn.preprocessing import StandardScaler

#create an instance of the scaler
scaler = StandardScaler()

#selecting the numerical columns
numerical_columns = ['age', 'fnlwgt', 'capital-gain', 'capital-loss', 'hours-per-week']

#normalise the selected columns
data[numerical_columns] = scaler.fit_transform(data[numerical_columns])
```
