```python
############### Write your code in this cell (If applicable) ##################

# Import necessary libraries
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
# Load the datasets
portugal = pd.read_csv('Portugal_online_retail.csv')
sweden = pd.read_csv('Sweden_online_retail.csv')
uk = pd.read_csv('UK_online_retail.csv')
```

```python
#display first five rows
portugal.head()
```

```python
#display first five rows
sweden.head()
```

```python
#display first five rows
uk.head()
```

>The invoice column won't be needed in the association as we proceed, hence we will drop it in each dataset

```python
#drop the invoice column
portugal.drop(columns='InvoiceNo',inplace = True)
sweden.drop(columns='InvoiceNo',inplace = True)
uk.drop(columns='InvoiceNo',inplace = True)
```

# Portugal Dataset

```python
#view the dataset after dropping inovice column
portugal.head()
```

```python
# Generate frequent itemsets from the Portugal dataset using a minimum support threshold of 10%. The output will include itemsets that are present in at least 10% of the transactions.

frq_items = apriori(portugal, min_support=0.1, use_colnames=True)
```

```python
#display the frq_items table

frq_items.head()
```

```python
#sort the dataset with the support column 

frq_items.sort_values(by=['support'],ascending=False)
```

```python
# Generate association rules based on confidence level of 70%

rules = association_rules(frq_items, metric="confidence", min_threshold=0.7)
```

```python
rules
```

```python
# Select the first three most important rules

top_three_rules = rules.head(3)
```

```python
top_three_rules
```

# TOP 3 RULES EXPLAINED
Rule 1: (BAKING SET 9 PIECE RETROSPOT) ➡ (RETROSPOT TEA SET CERAMIC 11 PC)

- People who buy the 9-piece Retrospot baking set often also buy the 11-piece Retrospot tea set.
- This suggests they might like having a matching kitchenware set.
- Stores could put these items together on shelves or recommend them online to make buying both easier.

Rule 2: (CHARLOTTE BAG SUKI DESIGN) ➡ (RED RETROSPOT CHARLOTTE BAG)

- People who buy the Charlotte Bag Suki Design also tend to buy the Red Retrospot Charlotte Bag.
- This means they probably like the Charlotte bag design a lot.
- Stores could offer deals where you get both bags together or target ads for both bags to people who bought one already.

Rule 3: (RED RETROSPOT CHARLOTTE BAG) ➡ (CHARLOTTE BAG SUKI DESIGN)

This rule is similar to rule 2 but the other way round.
- People who buy the red Charlotte bag might also want the Suki design one.
- This again shows customers like the Charlotte bag design and might collect both colors.
- Stores could promote these bags together to highlight that they come in different colors.
